### Adicionar 1 Coluna

```python
#a partir de uma coluna que existe
#vendas_df['Comissao'] = vendas_df['Valor Final']*0.05
#display(vendas_df)

#criar uma coluna com valor padrao
vendas_df.loc[:,'Imposto'] = 0  #no pandas ":" signica todos oss itens, no caso todas as linhas da coluna "Imposto" voa ser zeradas
display(vendas_df)
```

### Adicionar 1 linha
    - Linhas de um complemento da base de dados

```python
vendas_dez_df = pd.read_excel('Vendas - Dez.xlsx')
#display(vendas_dez_df)

vendas_df = pd.concat([vendas_df, vendas_dez_df])
#vendas_df['Comissao'] = vendas_df['Valor Final']*0.05 #Como naoa adicionei a coluna comissao na tabela dezembro os dados vem vazios
vendas_df['Imposto'] = 0 #o mesmo ocorre com a coluna impostos, tive que setar novamente para preencher os vazios
display(vendas_df)
```

### Excluir linhas e colunas

### Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
#deletar linhas e colunas completamente vazias
#vendas_df = vendas_df.dropna(how='all', axis=0)
#display(vendas_df)

#Deletar linhas com pelo menos 1 valor vazio
#vendas_df = vendas_df.dropna()

#preencher com media da coluna
#vendas_df['Comissao'] = vendas_df['Comissao'].fillna(vendas_df['Comissao'].mean())
#display(vendas_df)

#Preencher com o ultimo valor
vendas_df = vendas_df.ffill()
display(vendas_df)
```

### Calcular Indicadores
    - Groupby
    - Value Counts

```python
#quantas transacoes por loja?
#transacoes_loja = vendas_df['ID Loja'].value_counts()
#display(transacoes_loja)

#faturamento de cada produto
faturamento_prod = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_prod)
```

### Mesclar 2 dataframes (Procurar informações de um dataframe em outro)

```python
gerentes_df = pd.read_excel('Gerentes.xlsx') #importando tabela de gerentes
#display(gerentes_df)

#quero fazer um "procx" no vendas_df puxando o gerentes

vendas_df_gerentes = vendas_df.merge(gerentes_df) #Como em ambas as tabelas existe uma coluna com mesmo nome (ID Loja) ele automaticamente faz a conexao
display(vendas_df_gerentes)
```
